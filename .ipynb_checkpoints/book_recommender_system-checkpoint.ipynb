# Popularity Based Recommender System

```python
rating_with_name = ratings.merge(books,on='ISBN')
```

```python
books.head()
```

```python
rating_with_name
```

```python
num_rating_df = rating_with_name.groupby("Book-Title").count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={"Book-Rating":"num_rating"},inplace=True)
num_rating_df
```

```python
avg_rating_df = rating_with_name.groupby("Book-Title").mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={"Book-Rating":"avg_rating"},inplace=True)
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popularity_df
```

```python
popular_df =  popularity_df[popularity_df['num_rating']>250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
```

```python
popular_df
```

## Collaborative Filtering Based Recommender System

```python
x = rating_with_name.groupby("User-ID").count()['Book-Rating'] > 200
users = x[x].index
```

```python
filtered_rating = rating_with_name[rating_with_name['User-ID'].isin(users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50 
famous_book =  y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_book)]
```

```python
final_ratings.duplicated().sum()
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title',columns='User-ID',values = 'Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key = lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
    
    
```

```python
recommend('The Notebook')
```
